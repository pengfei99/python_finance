```python
import pandas as pd
import yfinance as yf
import numpy as np
```

```python

# read the wiki page of CAC 40, the fourth table contains the ticker infomation
ticker_df=pd.read_html("https://en.wikipedia.org/wiki/CAC_40")[4]
ticker_df.head()
```

```python
# get only tickers
ticker=ticker_df["Ticker"].to_list()
print(ticker)
```

```python
# The yfinance package can download all information about a company
# below we will get data of the CAC 40 since 2010. 
# "Adjusted close" is the closing price after adjustments for all applicable 
# splits and dividend distributions. 
df = yf.download(ticker,start="2010-01-01")['Adj Close']
```

```python
# The NAN means the stocks are not exchange back then. So we need to exlude the NAN
# for now we will drop the cloumn with NaN.

df.dropna(axis=1)
df.shape
```
